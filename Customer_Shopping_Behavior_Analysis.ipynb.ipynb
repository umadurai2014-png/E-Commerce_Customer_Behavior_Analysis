```python
# Checking if missing data or null values are present in the dataset
df.isnull().sum()
```

```python
# Impute missing Review Rating values with the median rating of their Category
df['Review Rating']= df['Review Rating'].fillna(
    df.groupby('Category')['Review Rating'].transform('median'))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation
df.columns= df.columns.str.lower()
df.columns= df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
lables = ['Young Adult','Adult','Middle-Age','Senior']
df['age_group']= pd.qcut(df['age'],q=4,labels= lables)
```

```python
df[['age','age_group']].head(20)
```

```python
# create new column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
# To not include the DataFrame’s index as a separate column
df.to_csv('customer_shopping_behaviour_cleaned_data.csv', index=False)
```
